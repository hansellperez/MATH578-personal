### Pure QR Algorithm

```python
def pure_qr(A, max_iter=50000):
    Ak = np.copy(A)
    n = A.shape[0]
    QQ = np.eye(n)
    for k in range(max_iter):
        Q, R = np.linalg.qr(Ak)
        Ak = R @ Q
        QQ = QQ @ Q
        if k % 100 == 0:
            print(Ak)
            print("\n")
    return Ak, QQ
```

### Practical QR Algorithm

```python
import numpy as np
from numpy.linalg import eig
from numpy.linalg import qr

def practical_qr(A, max_iter=5000):
    Ak = np.copy(A)
    n = A.shape[0]
    I = np.eye(n)
    for k in range(max_iter):
        sk, vk = eig(Ak[n-3:n-1,n-3:n-1])
        Qk, Rk = np.linalg.qr(Ak - min(sk)*I)
        Ak = Rk @ Qk + min(sk)*I
        if k % 100 == 0:
            print(Ak)
            print("\n")
    return Ak, Qk
```
